```python
import numpy as np
from scipy import sparse # 희소행렬(0을 많이 포함한 2차원 배열) 기능 제공
```

```python
eye = np.eye(5)
eye
```

```python
# numpy 배열을 CSR 포맷의 SciPy 희박 행렬로 변환
# 0이 아닌 원소만 저장
sparse_matrix = sparse.csr_matrix(eye)
print(sparse_matrix)
```

가로의 순서대로 재정렬하는 방법으로 행에 관여하여 정리 압축한 것을 CSR이라고 한다


보통 희소 행렬을 0이 모두 채워진 2차원 배열로 만들지 않음(메모리 부족 문제)

따라서 직접 만들 수 있어야함.

밑은 `COO`포맷으로 희소행렬 만들기

```python
data = np.ones(4) # 1으로 채워진 1x4 행렬 [1,1,1,1]
row_indices = np.arange(4) # [0,1,2,3]
# row_indices = np.array([1,1,1,1])
col_indices = np.arange(4) # [0,1,2,3]
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print(eye_coo)
```

## coo_matrix
- data: 어떤 값을 채울 건지
- row_indices: 행
- col_indices: 열
